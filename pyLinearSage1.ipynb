```python
# X = feature values, all the columns except the last column
X = newdf.iloc[:, [0,1,4,6]]
# y = target values, last column of the data frame
y = newdf.iloc[:, -1]
X.head()
```

```python
split = int(0.8*len(newdf)) # 80/29
#print(split)
X_train, X_test, Y_train, Y_test = X[:split], X[split:], y[:split], y[split:]
# Train    X  , Y
# Test     X,   Y
```

```python
# 74  , 59 = Train  , Test = 15
#1 , 2,3 ,4 ,-- 59, - TRain
#60,61. 62... 74  - test

#X[:split]  #[ : 59] # 0 to 59 # Train

#X[split:]  #[59:] # other than above

# 80/20 : 70/30
```

```python
import numpy as np

# LinearRegression
from sklearn.linear_model import LinearRegression

regModel = LinearRegression().fit(X_train, Y_train)

# FIT ( X,Y)
# FIT (X_TRAIN, Y_TRAIN)

# SKLEARN 
```

```python
regModel.score(X_train, Y_train)
```
