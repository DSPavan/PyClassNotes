```python
#isdigit is a method on strings, not integers.
#df2['AmountSpent']= df2.to_numeric(df2['AmountSpent'] , errors=coerce)

# find Errors
df2['AmountSpent'] = df2.drop(df2[df2['AmountSpent'].astype(float).isnull()].index)
df2.head()
```

```python
df2['AmountSpent'].convert_objects(convert_numeric=False)
df2.head()
```

```python
df2['AmountSpent'] = pd.to_numeric(df2['AmountSpent'], errors='coerce')
df2.head()
```

```python
df2['Gender']  = df2['Gender'].str[:2]
print(df2.head())  
```

```python
df2['Age']  = df2['Age'].str[:-2]
print(df2.head())  
```

```python
#Extract Last n characters from right of the column in pandas:
df2['Married']  = df2['Married'].str[-2:]
print(df2.head())  
```
